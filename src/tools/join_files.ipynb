```python
allFiles = glob.glob("data/*.csv")
# the header = 0 is important, because it kills the first row in the .csv that is the header. Do NOT user skiprows in param!
dfs = [pd.read_csv(f,header=0,error_bad_lines=False) for f in allFiles]
# ignore_index=True makes the indexing continuous instead of having reset each new join (1,2,3,4,...5000, vs 1,2,3...1,2,3...)
df = pd.concat(dfs, ignore_index=True) 
```
